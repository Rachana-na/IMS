```python
import json
fd = open("products.json",'r')
read = fd.read()
fd.close()
products = json.loads(read)
ui_product = input("Enter the product Id: ")
ui_quantity = int(input("Enter the quantity of the Product: "))
try:
    for prod in products.keys():
        if ui_quantity <= products[prod]['quantity']:
            print("****************************************")
            print("----------BILLING DETAILS:---------- ")
            print("product:",products[ui_product]['name'])
            print("price:",products[ui_product]['price'])
            print("Expiry date:",products[ui_product]['expiry'])
            print("Manufacturing Brand:",products[ui_product]['brand'])
            print("Total Amount: ",products[ui_product]['price']*ui_quantity) 
            print("**********************************")
            products[ui_product]["quantity"] = products[ui_product]["quantity"] - ui_quantity
        else:
            print("Sorry,the product is out of stock: ",products[ui_product]['quantity'])
        break
except:
    print("please enter valid product id")

```

### update products.json

```python
js = json.dumps(products)

fd = open("products.json",'w')
fd.write(js)
fd.close()
```
